```python
#3. Create a numpy array of ones and a numpy array of zeroes

import numpy as np
shape = (3,4)
array_of_ones = np.ones(shape)
print("Array_of_Ones(3*4):")
print(array_of_ones)

print("-"*16)

array_of_zeros = np.zeros(shape)
print("Array_of_Zeros(3*4):")
print(array_of_zeros)
```

```python
import numpy as np
shape = (6,3)
rand_6x3 = np.random.randint(0, 10, shape)

T_array = rand_6x3.T

print("Original 6x3 Array:")
print(rand_6x3)
print("\nTransposed Array (3x6):")
print(T_array)
```
